# Task 1: Discount Rules

```python
user_input = input("Enter the order amount: ")


# Check if input is numeric
if user_input.replace('.', '', 1).isdigit():
    order_amount = float(user_input)
    
    # Determine discount percentage
    if order_amount >= 2000:
        discount_percent = 15
    elif 1500 <= order_amount < 2000:
        discount_percent = 10
    elif 1000 <= order_amount < 1500:
        discount_percent = 7
    else:
        discount_percent = 0
        
    # Calculations
    discount_amount = order_amount * (discount_percent / 100)
    final_amount = order_amount - discount_amount
    
    print(f"Original Amount: {order_amount}")
    print(f"Discount: {discount_percent}% (-{discount_amount})")
    print(f"Final Total: {final_amount}")
else:
    print("Error: Invalid input. Please enter a numeric value.")
```

# Task 2: Process Multiple Orders

```python
orders = [1200, 2500, 800, 1750, 3000]
total_revenue = 0

print(f"{'Order Amount':<15} | {'Discount%':<10} | {'Final Amount':<15}")
print("-" * 45)

for amount in orders:
    if amount >= 2000:
        disc = 15
    elif 1500 <= amount < 2000:
        disc = 10
    elif 1000 <= amount < 1500:
        disc = 7
    else:
        disc = 0
        
    final = amount * (1 - disc/100)
    total_revenue += final
    print(f"{amount:<15} | {disc:<10} | {final:<15.2f}")

print("-" * 45)
print(f"Total Revenue after discounts: {total_revenue:.2f}")
```

# Task 3: User Menu

```python
order_list = []

while True:
    print("\n--- Menu ---")
    print("1 — Add order amount")
    print("2 — Show all orders and totals")
    print("q — Quit")
    
    choice = input("Choose an option: ").lower()
    
    if choice == 'q':
        print("Exiting menu...")
        break
    
    if choice == '1':
        val = input("Enter amount: ")
        if val.replace('.', '', 1).isdigit():
            order_list.append(float(val))
        else:
            print("Invalid number.")
            continue
            
    elif choice == '2':
        if not order_list:
            print("No orders recorded yet.")
        else:
            print(f"Orders: {order_list}")
            print(f"Total Count: {len(order_list)}")
    else:
        print("Invalid choice. Try again.")
        continue
```

# Task 4: Loop Control

```python
daily_sales = [200, 150, 0, 400, 50, -1, 300]
total_sales = 0

for sale in daily_sales:
    if sale == -1:
        print("Corrupted data found (-1). Stopping process.")
        break
    
    if sale == 0:
        print("No sales today, skipping...")
        continue
        
    total_sales += sale
    print(f"Added {sale}. Current total: {total_sales}")

print(f"Final Total Sales: {total_sales}")
```
