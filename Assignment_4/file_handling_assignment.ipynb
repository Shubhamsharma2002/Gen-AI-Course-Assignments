# Task 1: Write Sales Records to a File

```python
sales = [1200, 450, 980, 1500, 3000]

# Writing each sale on a new line
with open("sales_data.txt", "w") as file:
    for sale in sales:
        file.write(f"{sale}\n")

# Reopen and print contents
print("--- Task 1: File Contents ---")
with open("sales_data.txt", "r") as file:
    print(file.read())
```

# Task 2: Read File in Different Ways

```python
print("--- Task 2: Different Reading Methods ---")
with open("sales_data.txt", "r") as file:
    # 1. Read entire file
    # print(file.read()) 
    
    # 2. Read first line
    file.seek(0)
    print(f"First Line: {file.readline().strip()}")
    
    # 3. Readlines into a list of integers
    file.seek(0)
    sales_list = [int(line.strip()) for line in file.readlines()]
    print(f"Sales List: {sales_list}")
```

# Task 3: Append New Sales

```python
new_sales = [5000, 2500, 1700]
with open("sales_data.txt", "a") as file:
    for sale in new_sales:
        file.write(f"{sale}\n")

print("\n--- Task 3: Updated File ---")
with open("sales_data.txt", "r") as file:
    content = file.readlines()
    print("".join(content))
    print(f"Total lines: {len(content)}")
```

# Task 5: Create Product Info File (User Input)

```python
print("\n--- Task 5: User Input to File ---")
# Example input: Laptop, 1200; Mouse, 25; Keyboard, 45
with open("products.txt", "w") as file:
    for i in range(3):
        name = input(f"Enter product {i+1} name: ")
        price = input(f"Enter product {i+1} price: ")
        file.write(f"{name} | {price}\n")

print("\nProduct File Content:")
with open("products.txt", "r") as file:
    print(file.read())
```

# Task 6: Read File Safely (Error Handling)

```python
import os
print("--- Task 6: Safe File Reading ---")
filename = input("Enter filename to open: ")
if os.path.exists(filename):
    with open(filename, "r") as file:
        print(file.read())
else:
    print("File not found. Please check the filename.")
```

# Task 7: Mini Project - Export Discounted Prices

```python
prices = {
    "Mouse": 500,
    "Keyboard": 800,
    "Monitor": 7000,
    "Pendrive": 400,
    "Camera": 5000
}

discount_pct = float(input("\nEnter discount percentage (e.g., 10): "))
discounted_data = []

with open("discount_report.txt", "w") as file:
    file.write("Product | Original Price | Discounted Price\n")
    file.write("-" * 45 + "\n")
    for item, price in prices.items():
        disc_price = price - (price * discount_pct / 100)
        line = f"{item} | {price} | {disc_price:.2f}"
        file.write(line + "\n")
        discounted_data.append(disc_price)
    
    # Extra: Summary at the bottom
    file.write("-" * 45 + "\n")
    file.write(f"Total Items: {len(prices)}\n")
    file.write(f"Average Discounted Price: {sum(discounted_data)/len(discounted_data):.2f}\n")

print("\n--- Task 7: Discount Report Generated ---")
with open("discount_report.txt", "r") as file:
    print(file.read())
```
