# Task 1 - Basic Function: Price After Discount

```python
def apply_discount(price, discount_percent=5):
    """Returns the price after applying a discount (default 5%)."""
    # Extra: Ensure discount never exceeds 60%
    if discount_percent > 60:
        discount_percent = 60
    
    final_price = price - (price * (discount_percent / 100))
    return final_price


# Testing Task 1
print(f"Task 1: $1000 with 10% disc -> {apply_discount(1000, 10)}")
print(f"Task 1: $500 with default disc -> {apply_discount(500)}")
```

# Task 2 - Recursive Function: Factorial Utility

```python
def factorial(n):
    """Recursive function to find factorial of n."""
    if n < 0:
        return "Error: Negative numbers not allowed."
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)

# Testing Task 2
print(f"Task 2: Factorial(5) -> {factorial(5)}")
print(f"Task 2: Factorial(0) -> {factorial(0)}")
print(f"Task 2: Factorial(-3) -> {factorial(-3)}")
```

# Task 3 - Lambda Function: GST Calculator

```python
gst = lambda price: price + (0.18 * price)

# Extra: Final price after GST + Discount
gst_and_discount = lambda price, disc: (price - (price * disc/100)) * 1.18

print(f"Task 3: GST on 100 -> {gst(100)}")
```

# Task 4 - Using map(): Apply GST to List

```python
prices = [100, 250, 400, 1200, 50]
prices_with_gst = list(map(gst, prices))

print(f"Task 4: Original Prices -> {prices}")
print(f"Task 4: Prices after GST -> {prices_with_gst}")
```

# Task 5 - Using filter(): Filter Expensive Products

```python
prices_list = [100, 250, 400, 1200, 50, 2000, 850]
above_500 = list(filter(lambda p: p > 500, prices_list))
below_equal_500 = list(filter(lambda p: p <= 500, prices_list))

print(f"Task 5: Prices > 500 -> {above_500}")
print(f"Task 5: Prices <= 500 -> {below_equal_500}")
```

# Task 6 - Combined Utility Function

```python
def process_prices(prices_list):
    # Apply 10% discount to all
    discounted = list(map(lambda p: p * 0.9, prices_list))
    # Keep only those above 300
    filtered = list(filter(lambda p: p > 300, discounted))
    return discounted, filtered

# Testing Task 6
disc_all, final_filtered = process_prices([100, 500, 900, 50, 750])
print(f"Task 6: Discounted -> {disc_all}")
print(f"Task 6: Above 300 -> {final_filtered}")
```

# Task 7 - Mini Problem: Menu Using Functions

```python
def add_price(prices_list, price):
    prices_list.append(price)

def get_average_price(prices_list):
    return sum(prices_list) / len(prices_list) if prices_list else 0

def get_max_price(prices_list):
    return max(prices_list) if prices_list else 0

def run_menu():
    my_prices = []
    while True:
        print("\n--- Price Menu ---")
        print("1 -> Add price | 2 -> Show average | 3 -> Show highest | q -> Quit")
        choice = input("Choice: ").lower()
        
        if choice == 'q': break
        elif choice == '1':
            p = float(input("Enter price: "))
            add_price(my_prices, p)
        elif choice == '2':
            print(f"Average: {get_average_price(my_prices)}")
        elif choice == '3':
            print(f"Highest: {get_max_price(my_prices)}")
        else:
            print("Invalid Choice.")
 # Uncomment to run interactively
# run_menu()   
```
